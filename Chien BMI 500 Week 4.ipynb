```python
"""
Frank Chien
BMI 500 Homework Week 4
Using an unsupervised clustering algorithim and applying it to the breast cancer dataset with scitkit-learn
This homework applies kmeans using varius number of clusters to group the Wisconsin breast cancer dataset
The dataset uses digitized images to compute 10 attributes per sample: radius, texture, area, etc.. 
and has a corresponding 'benign' vs 'malignant' classification
"""

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import cluster, datasets

#loading breast cancer data
breast_cancer_bunch = datasets.load_breast_cancer(as_frame=True)
breast_cancer_df = breast_cancer_bunch["frame"] #extracts the dataframe from bunch

#this will plot radius by texture

plt.scatter(breast_cancer_df["mean radius"],breast_cancer_df["mean texture"])
plt.title("Breast Cancers by Radius and Texture")
plt.show()
plt.savefig("scatter_uncolored.png",bbox_inches='tight')
plt.close('all')
```

```python
#exploring our dataframe
print(breast_cancer_df)
print(breast_cancer_df.columns)
print(breast_cancer_bunch["DESCR"])
```

```python
#plots the scatter plot with colors based on malignant/benign
plt.scatter(breast_cancer_df["mean radius"],breast_cancer_df["mean texture"],c=breast_cancer_df["target"])
plt.title("Breast Cancers by Radius and Texture, Malignant vs Benign Groupings")
plt.show()
plt.close('all')

```

```python
"""
now to apply kmeans
"""
#start with 2 clusters
k_means_2 = cluster.KMeans(n_clusters=2)
sub_df = breast_cancer_df[["mean radius","mean texture"]] #looks only at the first 2 attributes
target_2 = k_means_2.fit_predict(sub_df) #trains and predicts 2 clusters based on first 2 attributes

#adding 2-cluster predicted targets to database
breast_cancer_df["predicted2"] = target_2

#plots the scatter plot with new targets based on kmeans2
plt.scatter(breast_cancer_df["mean radius"],breast_cancer_df["mean texture"],c=breast_cancer_df["predicted2"])
plt.title("Breast Cancers by Radius and Texture, predicted Targets")
plt.show()
plt.close('all')
```

```python
#now to do kmeans with additional clusters
#training kmean with 4 clusters 

k_means_4 = cluster.KMeans(n_clusters=4)
target_4 = k_means_4.fit_predict(sub_df)

#adding targets to df
breast_cancer_df["predicted4"] = target_4

#and plotting
plt.scatter(breast_cancer_df["mean radius"],breast_cancer_df["mean texture"],c=breast_cancer_df["predicted4"])
plt.title("Breast Cancers by Radius and Texture, predicted Targets")
plt.show()
plt.close('all')
```

```python
k_means_8 = cluster.KMeans(n_clusters=8)
target_8 = k_means_8.fit_predict(sub_df)

#adding targets to df
breast_cancer_df["predicted8"] = target_8

#and plotting
plt.scatter(breast_cancer_df["mean radius"],breast_cancer_df["mean texture"],c=breast_cancer_df["predicted8"])
plt.title("Breast Cancers by Radius and Texture, predicted Targets")
plt.show()
plt.close('all')
```
